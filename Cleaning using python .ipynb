# Cleaning individual columns


```python
# Cleaning Column 'Host_identity_verified' 
print(df['Host_identity_verified'].unique())
print(df['Host_identity_verified'].value_counts())
df['Host_identity_verified']=df['Host_identity_verified'].apply(lambda x : x .strip().capitalize())
df
```

```python
# Add new column near to Instant_bookable column
df.insert(
    df.columns.get_loc('Instant_bookable') + 1,  # Position next to 'Instant_bookable'
    'Binary_Instant_bookable',  
    df['Instant_bookable'].apply(lambda x: 1 if x == True else 0)  # Convert to binary
)
```

# check for some more Transformations .


```python
# [ price column ]  is object type     ==> need to convert to int and  remove $ sign and remove sapces
df['Price'].dtype
df['Price']=df['Price'].str.replace('$','').str.replace(',','').astype(int)
```

```python
# [service fee]  is object type     ==> need cleaning to float and  remove $ sign
print(df['Price'].head())
```

```python
df['Service_fee']=df['Service_fee'].str.replace("$","").str.replace(",","").str.replace(" ","").astype(float)
```
